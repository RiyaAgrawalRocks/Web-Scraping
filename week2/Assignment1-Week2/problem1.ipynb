```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('NDAP_REPORT_6000.csv', sep=',')
```

```python
#print(data.info())
new_data=data.dropna()
print(new_data.info())
```

```python
# We see that there are no empty cells
new_data.drop_duplicates(inplace=True)
print(new_data.info())
```

```python
# checking for null values or values of different data type
for column in new_data.columns :
    if new_data[column].isnull().any():
        print(column + 'has null value(s)')
    a=len(new_data[column].map(type).unique())
    if a!=1:
        print(column + 'has more than one dtype')

# finding the literacy rates
#new_data['Total literacy rate']=new_data['Literate population']*100/new_data['Population']
new_data['Male literacy rate']=new_data['Male literate population']*100/new_data['Male population']
#new_data['Female literacy rate']=new_data['Female literate population']*100/new_data['Female population']

# plotting the literacy rates to find outliers
new_data=new_data.groupby('State')
for state, group in new_data:
#    plt.subplot(1,2,1)
 #   plt.scatter(group['Total literacy rate'])
    plt.subplot(1,2,2)
    plt.scatter(group['Male literacy rate'])
#    plt.subplot(1,2,3)
#    plt.scatter(group['Female literacy rate'])
```

```python
lit_rates=new_data.groupby(['State','Rural or Urban']).agg(
             total_lit = pd.NamedAgg(column='Literate population' , aggfunc=lambda x : x['Literate population']/x['Population'])
)
print(lit_rates)
""" 'female_lit': (lambda x : x['female literate population']/x['Female population']),
             'male_lit': (lambda x : x['Male literate population']/x['Male population']) """
```
